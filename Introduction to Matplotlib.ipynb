## An example workflow

```python
# 1 import lib
%matplotlib inline
import matplotlib.pyplot as plt
#2 prepare the data
x = [10,20,30,45]
y = [100,250, 60,80]
#3 create the figure and subplot
fig, ax = plt.subplots(figsize = (10,10))
ax.plot(x,y)
ax.set(title = 'sample',
        xlabel = 'x axis',
        ylabel = 'y axis')
#4 output
plt.show()
fig.savefig('images\sample-matplotlib.png')
```

# Making figures with numpy array
We want:
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplot

```python
import numpy as np
```

```python
#create some data
x = np.linspace(0,10,100)
```

```python
#plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x,x**2);
```

```python
# Use same data to make a scatter plot
fig, ax = plt.subplots()
ax.scatter(x,np.exp(x));
```

```python
# Use same data to make a scatter plot
fig, ax = plt.subplots()
ax.scatter(x,np.sin(x));
```

```python
# Make a plot from dictionary
nut_butter_prices = {"Almond butter":10,
                    "Peanut butter":8,
                    "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values());
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()));
```

```python
#make some data
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

## Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2),(ax3, ax4)) = plt.subplots(nrows = 2,
                                           ncols = 2,
                                           figsize=(10,5))
#Plot to each different axis
ax1.plot(x,x/2)
ax2.plot(x,x**2)
ax3.plot(x,x**3)
ax4.plot(x,2*(x));
```

```python
# option 2
fig, ax = plt.subplots(nrows = 2,
                        ncols = 2,
                        figsize=(10,5))
#plot to each differnt index
ax[0,0].plot(x,x/2);
```

## Ploting from pandas dataframe

```python
car_sales = pd.read_csv('https://raw.githubusercontent.com/xtructt/z2m-ML/master/Resources/zero-to-mastery-ml-master/data/car-sales.csv')
car_sales
```

```python
ts = pd.Series(np.random.randn(1000),
              index = pd.date_range("1/1/2020",periods=1000))
ts = ts.cumsum()
ts.plot();
```

```python
car_sales["Price"] = car_sales["Price"].str.replace("[\$\,\.]","")
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].str[:-2]
```

```python
car_sales["Sale Date"] = pd.date_range("1/1/2020",periods=len(car_sales))
```

```python
car_sales['Total Sales'] = car_sales["Price"].astype(int).cumsum()
```

```python
car_sales
```

```python
#plot the total sales
car_sales.plot(x="Sale Date", y = "Total Sales")
```
