## An example workflow

```python
# 1 import lib
%matplotlib inline
import matplotlib.pyplot as plt
#2 prepare the data
x = [10,20,30,45]
y = [100,250, 60,80]
#3 create the figure and subplot
fig, ax = plt.subplots(figsize = (10,10))
ax.plot(x,y)
ax.set(title = 'sample',
        xlabel = 'x axis',
        ylabel = 'y axis')
#4 output
plt.show()
fig.savefig('images\sample-matplotlib.png')
```
